```python
import osmnx as ox
```

```python
%matplotlib inline
```

```python
ox.config(log_file=True, log_console=True, use_cache=True)
```

```python
city = ox.gdf_from_place('Zurich, Zurich, Switzerland')
```

```python
city
```

```python
ox.save_gdf_shapefile(city)
```

```python
city_pj = ox.project_gdf(city)
```

```python
ox.plot(city_pj)
```

```python
fig, ax = ox.plot_shape(city_pj)
```

```python
g = ox.graph_from_place('Zurich, Zurich, Switzerland')
```

```python
g_pj = ox.project_graph(g)
fig, ax = ox.plot_graph(g_pj)
```

```python
v = ox.graph_from_place('Volketswil, Zurich, Switzerland', network_type='walk')
```

```python
v_pj = ox.project_graph(v)
fig, ax = ox.plot_graph(v_pj, node_size=30, node_color='#66cc66')
```

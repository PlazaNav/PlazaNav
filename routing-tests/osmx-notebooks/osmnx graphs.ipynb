```python
import osmnx as ox
%matplotlib inline
ox.config(log_file=True, log_console=True, use_cache=True)
```

```python
# helvetiaplatz, Zurich
north, east, south, west = 47.37666, 8.52797, 47.37544, 8.52560
```

```python
g = ox.graph_from_bbox(north, south, east, west, network_type='walk')
```

```python
proj = ox.project_graph(g)
fig, ax = ox.plot_graph(proj)
```

```python
m = ox.graph_to_gdfs(g)
m. = 'test'
m_proj = ox.project_gdf(m)
fig, ax = ox.plot(m_proj)
```
